```python
d = 150
maps = [[0, 50, 10], [0, 50, 20], [50, 100, 10], [100, 151, 10], [110, 140, 90]]
```

```python
position = [i for i in range(d+1)]
```

```python
for j in range(d+1):
    if j != 0:
        position[j] = min(position[j],position[j-1]+1)
    for i in maps:
        if i[0] == j and i[1] <= d:
            position[i[1]] = min(position[i[1]],position[i[0]]+i[2])
```

```python
print(position[d])
```
