```python
input = '5 17' # input 값 입력
```

```python
n, k = map(int,input.split()) # 출발 지점, 도착 지점 변수 값 지정
```

```python
Max = 10000
```

```python
dis = [0] * Max
```

```python
def bfs(n,k,Max): # BFS 함수 정의
    queue = [] 
    queue.append(n) # 출발 지점 Node에 입력
    
    while queue:
        x = queue.pop(0)
        if x == k: # 출발지점에서 이동한 Node가 
            return dis[x]
            
            
        for j in (x-1, x+1, x*2):
            if 0 <= j <= Max and not dis[j]:
                dis[j] = dis[x] + 1
                queue.append(j)
```

```python
bfs(n,k,Max)
```

**아래의 그림과 같이 모든 node에 대한 탐색을 통해서 최소점을 찾는 방법으로 문제를 해결하였다.**
