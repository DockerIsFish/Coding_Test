```python
n = 5
height = [[6, 8, 2, 6, 2],
[3, 2, 3, 4, 6],
[6, 7, 3, 3, 2],
[7, 2, 5, 3, 6],
[8, 9, 5, 2, 7]]
```

```python
def bfs (start, safe_area):
    
    queue = []
    a, b = start
    queue.append((a,b))
    visited[a][b] = 1

    while queue:

        dx = [1,-1,0,0]
        dy = [0,0,1,-1]
        x,y = queue.pop(0)
        for i in range(4):

            nx = x + dx[i]
            ny = y + dy[i]

            if nx < 0 or nx >= n or ny <0 or ny >= n:
                continue

            if height[nx][ny] >= safe_area and visited[nx][ny] == 0:
                visited[nx][ny] = 1
                queue.append((nx,ny))
```

```python
graph_min = min(map(min, height))    
graph_max = max(map(max, height))
max_safe_area = graph_min
for safe_area in range(graph_min, graph_max+1):
    visited = [[0] * n for _ in range(n)]
    temp = 0
    for i in range(n):
        for j in range(n):
            if height[i][j] >= safe_area and visited[i][j] == 0:
                start = (i,j)
                bfs(start, safe_area)
                temp += 1
                
    if temp > max_safe_area:    # update safe_zone count
        max_safe_area = temp
```
