```python
n,m = 7,6
```

```python
graph = [[1, 2],
[2, 3],
[1, 5],
[5, 2],
[5, 6],
[4, 7]]
```

```python
answer = []
```

```python
def bfs (start):
    
    queue = []
    queue.append(start)
    visited = [False] * (n + 1)
    visited[start] = True
    while queue:
        
        x = queue.pop(0)
        
        for node in graph:
            if node[0] == x and visited[node[1]] == False:
                queue.append(node[1])
                visited[node[1]] = True
                answer.append(node[1])
        print(queue)
```

```python
bfs(1)
```
