```python
n = 4
```

```python
answer = []
for num in range(10**(n-1),10**(n)):
    if check(num):
        num_str = str(num)
        #print(num_str)
        count = 0
        for i in range(1,n):
            num_int = int(num_str[0:i])
            #print( num_int,'*')
            if check (num_int) and num_int != 1:
                #print('소수')
                count += 1
        if count == n-1:
            answer.append(num)
```

```python
def check (num):
    i = 2
    while i * i <= num:
        if num % i == 0:
            return False
        i += 1
    return True
```

```python
print(answer)
```

제곱근까지 나눠서 나눠 떨어뜨리는 수가 존재하면 해당 수는 소수가 아니다

```python
def find(num):
    if num < 2:
        return False
    for i in range(2, int(int(num) ** 0.5) + 1):
        if int(num) % i == 0:
            return False
    
    return True

def dbs(num): # 목표 길이 도달 시 멈춤
    
    if len(str(num)) == n :
        print(num)
        
    else:
        for i in range(10):
            temp = num + 10*i
            if find(temp):
                dfs(temp)
```
