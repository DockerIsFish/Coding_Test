```python
n,m = 4, 2
```

```python
s = []
answer = []
```

```python
def dfs(start):
    
    if len(s) == m:
        answer.append(''.join(map(str,s)))
        return 
    
    for i in range(start,n+1):
        if i not in s:
            s.append(i)
            dfs(i+1)
            s.pop()
```

```python
dfs(1)
```

```python
answer
```

재귀함수를 이용하고 출력하는 리스트의 개수가 m 개이면 리스트의 내용을 출력
재귀함수에서는 전달받은 인자인 start부터 n까지의 수자를 사용
재귀함수를 호출할 때에는 i + 1을 호출
다음에 오는 숫자가 현재 숫자보다 작으면 가지치기하여 검색하지 않는다.
