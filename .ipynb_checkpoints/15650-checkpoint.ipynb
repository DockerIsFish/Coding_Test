```python
n,m = 4, 2
```

```python
s = []
answer = []
```

```python
def dfs(start):
    
    if len(s) == m:
        answer.append(''.join(map(str,s)))
        return 
    
    for i in range(start,n+1):
        if i not in s:
            s.append(i)
            dfs(i+1)
            s.pop()
```

```python
dfs(1)
```
