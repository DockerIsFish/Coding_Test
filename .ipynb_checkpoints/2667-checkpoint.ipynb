```python
input = '0110100011010111101010000111010000001111100111000' 
n = 7
```

```python
arr = []
for i in range(n):
    pre_arr = [ int(i) for i in input[i*n:i*n+n]]
    arr.append(pre_arr)
```

```python
def bfs(arr,a,b):
    dx = [0,0,1,-1]
    dy = [1,-1,0,0]
    n = len(arr)
    queue = []
    queue.append((a,b))
    arr[a][b] = 0 
    count = 1 
    
    while queue:
        x,y = queue.pop(0)
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if nx <0 or nx >=n or ny <0 or ny>=n:
                continue
            if arr[nx][ny] == 1:
                arr[nx][ny] = 0 
                queue.append((nx,ny))
                count +=1
                
    return count
```

```python
cnt = []
for i in range(n):
    for j in range(n):
        if arr[i][j] == 1:
            cnt.append(bfs(arr,i,j))
```

```python
cnt.sort()
```

```python
answer = [len(cnt)] + cnt
```

```python
print(answer)
```
