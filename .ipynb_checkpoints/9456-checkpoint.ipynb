```python
def bfs(arr,start):
    len_x = len(arr[0])
    len_y = len(arr)
    visited =[[0]*len_x]*len_y
    queue = []
    answer = []
    a,b = start 
    queue.append((b,a))
    
    while queue:
        dx = [1,-1,0,0]
        dy = [0,0,1,-1]
        y,x = queue.pop(0)
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if nx < 0 or nx >= len_x or ny < 0 or ny >= len_y:
                continue
            if visited[ny][nx] == 0: #    
                if visited[y][x] == 0: #     
                    visited[ny][nx] = 1
                    answer.append(arr[ny][nx])
                   # print(ny,nx,arr[ny][nx])
                queue.append((ny,nx))
    
    return answer
```
