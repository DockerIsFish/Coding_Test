```python
n,m,start = 5,6,1
input = [[5, 1, 1],[1, 2, 2],[1, 3, 3],[2, 3, 4],[2, 4, 5],[3, 4, 6]]
```

```python
graph = [ [] for i in range(n+1)]   
visited = [False] * (n+1)
distance =[INF] * (n+1)
```

```python
INF = 1e9
```

```python
for i in input:
    graph[i[0]].append((i[1],i[2]))
```

```python
def get_samllest_node():
    min_value = INF
    index = 0 
    for i in range(1,n+1):
        if not visited[i] and distance[i] < min_value:
            min_value = distance[i]
            index = i 
    return index    
```

```python
def dijkstra(start):
    #    0  
    distance[start] = 0 
    visited[start] = True 
    #     
    for i in graph[start]:
        distance[i[0]] = i[1]
    #    n-1  
    for _ in range(n-1):
        now = get_samllest_node() #    
        visited[now] = True #    
        #     
        for next in graph[now]:
            cost = distance[now] + next[1]
            if cost < distance[next[0]]:
                distance[next[0]] = cost
```

```python
dijkstra(start)
```

```python
for i in range(1, n+1):
    if distance[i] == INF:
        print('INF')
    else:
        print(distance[i])
```
