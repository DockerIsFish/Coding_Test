```python
n, m = 3, 2
```

```python
arr = [[1,3],[2,3]]
```

```python
inDegree = [0 for i in range(32001)] # 진입 차수 
```

```python
graph = [[] for i in range(32001)] # 
```

```python
queue = []
```

```python
for a,b in arr:
    inDegree[b] += 1
    graph[a].append(b)
```

```python
for i in range(1,n+1):
    if inDegree[i] == 0:
        queue.append(i)
```

```python
while queue:
    student = queue.pop(0)
    for j in graph[student]:
        inDegree[j] -=1
        if inDegree[j] == 0:
            queue.append(j)
    print(student, end = ' ')
```

A번 학생이 B번 학생보다 반드시 먼저 앞에와야 하는 상황
A -> B 를 만족시키면서 정렬을 해야한다. 

위상정렬의 과정

- 진입차수 0 인 정점을 큐에 삽입
- 큐에서 원소를 꺼내 해당 원소와 연결된 간선을 모두 제거
- 제거한 후에 진입차수가 0 인 정점을 큐에 삽입
- 2 번째와 3 번째 과정을 반복
