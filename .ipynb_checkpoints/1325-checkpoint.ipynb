```python
input = [[3,1],[3, 2],[4, 3],[5, 3]]
```

```python
n = 5
```

```python
arr =  [ [] for i in range(0,n+1)]
```

```python
for a,b in input:
    arr[b].append(a)
#    if a not in arr[b]:
#        arr[b].append(a)
```

```python
arr
```

```python
def bfs(arr,start):
    cnt = 1
    visited = [False] * (n+1)
    visited[start] = True
    queue = []
    queue.append(start)
    while queue:
        v = queue.pop(0)
        #print(v)
        for i in arr[v]:
            if not visited[i]:
                #print(i,'**')
                queue.append(i)
                visited[i] = True
                cnt += 1
    return cnt#,visited
```

```python
bfs(arr,1)
```

```python
results = []
max_cnt = 0
for i in range(1,n+1):
    cnt = bfs(arr,i)
    #print(cnt)
    if cnt > max_cnt:
        max_cnt = cnt
    
    results.append([i,cnt])
```

```python
max = 0
answer = []
for result in results:
    if max <= result[1]:
        max = result[1]
        answer.append(result[0])
```

간단한 BFS 문제이다. 하지만 양방향이 아니라 단방향이라는 것에 중의해야한다. 
그리고 start 노드를 방문처리 하는 것을 잊지 말자.
