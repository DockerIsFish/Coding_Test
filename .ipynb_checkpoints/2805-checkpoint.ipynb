```python
n,m = 5,20
```

```python
trees = [4,42,40,26,46]
```

```python
def find (height):
    ans = 0
    for tree in trees:
        if tree > height:
            ans += tree - height
    return ans            
```

```python
for i in range(1000):
    if m == find(i):
        print(i)
```

```python
find(36)
```

```python
start, end = 0, max(trees)
```

```python
while start <= end:
    mid = ( start + end ) // 2
    ans = 0 
    for tree in trees:
        if tree > mid:
            ans += tree - mid
    if ans >= m:
        start = mid + 1
    else:
        end = mid - 1
print(end)
```

이분 탐색을 활용한 알고리즘 문제 풀이
1. 가장 큰 나무와 작은 나무의 중간 지점을 톱의 높이로 지정
2. 중간지점의 톱의 높이를 기준으로 필요한 나무보다 많은면 start +1 작으면 end -1 
3. start 가 end 보다 커질결우까지 계속 이분 탐색을 진행한다.
