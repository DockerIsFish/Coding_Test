```python
n,m = 4,6
```

```python
input = '101111101010101011111011'
```

```python
arr = []
for i in range(n):
    pre_arr = [int(i) for i in input[m*i:m*i+m]]
    arr.append(pre_arr)
```

```python
arr
```

```python
def bfs(arr,a,b):
    dx = [-1,1,0,0]
    dy = [0,0,1,1]
    
    queue = []
    queue.append((a,b))
    
    while queue:
        x,y = queue.pop(0)
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny <0 or ny >= m:
                continue
            
            if arr[nx][ny] == 0:
                continue
                
            if arr[nx][ny] == 1:
                arr[nx][ny] = arr[x][y] + 1
                queue.append((nx,ny))
                
    return arr[n-1][m-1]
```

```python
bfs(arr,0,0)
```

BFS 알고리즘을 활용한 풀이
** continue, break, pass 에 대한 정확도 이해가 필요
