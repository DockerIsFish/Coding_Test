```python
n = 7
input = '10 10 10 30 10 40 50'
```

```python
a=  list(map(int,input.split()))
dp = [0 for i in range(n)]
for i in range(n):
    for j in range(i):
        if a[i] > a[j] and dp[i] < dp[j]:
            dp[i] = dp[j]
    dp[i] += 1
print(max(dp))
```

DP 알고리즘을 활용하여 각 숫자가 자기 전의 모든 숫자보다 크면 +1, 그 외는 자기와 같은 숫자의 길이를 저장하게끔 설정하여 문제를 해결하였다.
