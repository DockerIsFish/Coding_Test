# 이진트리 순회(DFS)

---
아래 그림과 같은 이진트리를 전위순회와 후위순회를 연습해보세요.

전위순회 출력 : 1 2 4 5 3 6 7

중위순회 출력 : 4 2 5 1 6 3 7

후위순회 출력 : 4 5 2 6 7 3 1

---
## 코드

```python
# 전위 순회
def DFS(x):
    if x>7:
        return
    else:
        print(x, end=" ")
        DFS(x*2)
        DFS(x*2+1)

if __name__=="__main__":
    n=int(input())
    DFS(n)
```

위의 순서로 작동한다.

```python
# 중위 순회
def DFS(x):
    if x>7:
        return
    else:
        DFS(2*x)
        print(x, end=" ")
        DFS(2*x+1)

if __name__=="__main__":
    n=int(input())
    DFS(n)
```

```python
# 후위 순회
def DFS(x):
    if x>7:
        return
    else:
        DFS(2*x)
        DFS(2*x+1)
        print(x, end=" ")

if __name__=="__main__":
    n=int(input())
    DFS(n)
```
