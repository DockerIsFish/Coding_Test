# 중복순열 구하기(DFS)
---
1부터 N까지 번호가 적힌 구슬이 있습니다. 

이 중 중복을 허락하여 M번을 뽑아 일렬로 나열하는 방법을 모두 출력합니다.

---
### 입력설명
첫 번째 줄에 자연수 N(3<=N<=10)과 M(2<=M<=N) 이 주어집니다.

---
### 출력설명
첫 번째 줄에 결과를 출력합니다. 맨 마지막 총 경우의 수를 출력합니다.
출력순서는 사전순으로 오름차순으로 출력합니다.

---
### 입력예제 1
3 2

---
### 출력예제 1
1 1

1 2

1 3

2 1

2 2

2 3

3 1

3 2

3 3

9

---
## 코드

```python
def dfs(L):
    global count
    
    if L==m: # 하나의 중복순열 완성
        for j in range(m):
            print(subset[j], end=" ")
        print()
        count+=1

    else:
        for i in range(1, n+1): # 1부터 n까지
            subset[L]=i         # L번째에 i 넣기
            dfs(L+1)            # 한칸 내려가기


if __name__=="__main__":
    n, m = map(int, input().split()) # n까지 m번 뽑기
    subset = [0]*m # 부분집합 초기화
    count = 0 
    dfs(0)
    print(count)
    
# 스택에 dfs(0) - dfs(1) - dfs(2) - pop - dfs(2) - pop - dfs(2) - pop - pop - dfs(2)... 저장된다
#                     dfs(0)
#                  1/   2|    \3
#            dfs(1)    dfs(2)  dfs(3)
#         1/   2|  3\
#    dfs(2)  dfs(2) dfs(2)
#    (1,1)   (1,2)  (1,3)  (2,1) (2,2) (2,3) (3,1) (3,2) (3,3)
```
