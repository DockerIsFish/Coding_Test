```python
n,m = 5,20
```

```python
trees = [4,42,40,26,46]
```

```python
def find (height):
    ans = 0
    for tree in trees:
        if tree > height:
            ans += tree - height
    return ans            
```

```python
for i in range(1000):
    if m == find(i):
        print(i)
```

```python
find(36)
```

```python
start, end = 0, max(trees)
```

```python
while start <= end:
    mid = ( start + end ) // 2
    ans = 0 
    for tree in trees:
        if tree > mid:
            ans += tree - mid
    if ans >= m:
        start = mid + 1
    else:
        end = mid - 1
print(end)
```
