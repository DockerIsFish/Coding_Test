```python
n,m = 5 ,4
input = '11101101011010110111'
```

```python
arr = []
for i in range(m):
    arr.append([i for i in map(int,input[i*n:i*n+n])])
```

```python
arr
```

```python
def bfs(a,b,arr,m,n):
    queue = []
    queue.append((a,b))
    
    while queue:
        dx = [-1,0,1,-1,1,-1,0,1] 
        dy = [1,1,1,0,0,-1,-1,-1]
        x, y = queue.pop(0)
        for i in range(8):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 > nx or nx >= n or ny < 0 or ny >= m:
                continue
            
            if arr[nx][ny] == 1:
                arr[x][y] = 0
                queue.append((nx,ny))
```

```python
for j in range(n):
    for i in range(m):
        bfs(i,j,arr,m,n)
```
