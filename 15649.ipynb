백트래킹

DFS ( 깊이 우선 탐색 )의 방식을 기반으로, 불필요한 경우를 배제하며 원하는 해답에 도달할 때 까지 탐색하는 전략

```python
n, m = 4,2
```

```python
s = []
```

```python
def dfs():
    if len(s) == m:
        print(' '.join(map(str,s)))
        return
    
    for i in range(1,n+1):
        if i in s:
            continue
        s.append(i)
        dfs()
        s.pop()
```

```python
dfs()
```
