```python
grades = [[3,6],[7,3],[4,2],[1,4],[5,7],[2,5],[6,1]]
```

```python
arr = [0] * (len(grades))
```

```python
arr
```

```python
for grade in grades:
    arr[grade[0]-1] = grade[1]
```

```python
arr
```

```python
stack = []
count = 0
for idx in arr:
    
    if bool(stack) == False:
        stack.append(idx)
        count +=1
    if idx < stack[0]:
        count += 1
        stack.pop()
        stack.append(idx)
```
