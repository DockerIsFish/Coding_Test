```python
input = '5 17' # input  
```

```python
n, k = map(int,input.split()) #  ,     
```

```python
Max = 10000
```

```python
dis = [0] * Max
```

```python
def bfs(n,k,Max): # BFS  
    queue = [] 
    queue.append(n) #   Node 
    
    while queue:
        x = queue.pop(0)
        if x == k: #   Node 
            return dis[x]
            
            
        for j in (x-1, x+1, x*2):
            if 0 <= j <= Max and not dis[j]:
                dis[j] = dis[x] + 1
                queue.append(j)
```

```python
bfs(n,k,Max)
```
