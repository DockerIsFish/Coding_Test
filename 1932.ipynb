```python
n = 5 

input = '738810274445265'
input= [i for i in map(int,input)]
```

```python
arr = []
for i in range(n):
    pre_arr = []
    for j in range(i+1):
        pre_arr.append(input.pop(0))
    arr.append(pre_arr)
```

```python
for i in range(1,n):
    for j in range(len(arr[i])):
        if j == 0:
            arr[i][j] += arr[i-1][j] 
        elif j == i:
            arr[i][j] += arr[i-1][j-1]
        else: 
            arr[i][j] += max(arr[i-1][j],arr[i-1][j-1])
```

```python
answer = max(arr[n-1])
```
