```python
graph = [[0, 3, 5, 4, 6, 9, 2, 7, 8],
[7, 8, 2, 1, 0, 5, 6, 0, 9],
[0, 6, 0, 2, 7, 8, 1, 3, 5],
[3, 2, 1, 0, 4, 6, 8, 9, 7],
[8, 0, 4, 9, 1, 3, 5, 0, 6],
[5, 9, 6, 8, 2, 0, 4, 1, 3],
[9, 1, 7, 6, 0, 2, 0, 8, 0],
[6, 0, 3, 7, 0, 1, 9, 5, 2],
[2, 5, 8, 0, 0, 4, 7, 6, 0]]
```

```python
blank = []
```

```python
for i in range(9): # 0 의 좌표를 blank List에 저장
    for j in range(9):
        if graph[i][j] == 0:
            blank.append((i,j))
```

```python
def checkRect(x,y,a): # 3*3 정사각형에 없는 수임을 확인 
    nx = x//3 * 3 
    ny = y//3 * 3 # 몇 번째 칸에 있는지 확인 가능 
    
    for i in range(3):
        for j in range(3):
            if a == graph[nx+i][ny+i]:
                return False
    return True
```

```python
def checkRow(x,a):
    for i in range(9):
        if a == graph[x][i]:
            return False 
    return True
```

```python
def checkCol(y,a):
    for i in range(9):
        if a == graph[i][y]:
            return False
    return True
```

```python
def dfs(idx):
    if idx == len(blank):
        for i in range(9):
            print(graph[i])
        return 
    for i in range(1,10):
        x = blank[idx][0]
        y = blank[idx][1]
        
        if checkRow(x,i) and checkCol(y,i) and checkRect(x,y,i):
            #print(x,y,i)
            graph[x][y] = i 
            dfs(idx + 1)
            graph[x][y] = 0
```
